```python
dataset= pd.read_csv("D:/Besant Technologies/Python_Data_Science/Datasets/Instagram visits clustering.csv")
```

```python
dataset.head(10)
```

```python
X= dataset.iloc[:, [1,2]].values
```

```python
plt.figure(figsize=(10,8))

plt.scatter(X[:, 0], X[:, 1], s=50)

plt.xlabel("IG Visit Score")

plt.ylabel("Spending Rank")

plt.title("IG Visit History")

plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss= []

for i in range(1, 11):

  kmeans= KMeans(n_clusters= i, init= "k-means++", random_state= 42)

  kmeans.fit(X)

  wcss.append(kmeans.inertia_)

print(wcss)
```

```python
#elbow curve


plt.figure(figsize=(8,6))

plt.plot(range(1,11), wcss, marker= "o", linestyle= "--")

plt.xlabel("Number of clusters")

plt.ylabel("WCSS")

plt.show()
```

```python
kmeans= KMeans(n_clusters= 5, init= "k-means++", random_state=42)

y_kmeans= kmeans.fit_predict(X)

print(y_kmeans)
```

```python
plt.figure(figsize= (10,8))

plt.scatter(X[y_kmeans==0,0], X[y_kmeans==0,1], s=100, c= "red", label= "Cluster 1")

plt.scatter(X[y_kmeans==1, 0], X[y_kmeans==1, 1], s=100, c= "blue", label= "Cluster 2")

plt.scatter(X[y_kmeans==2, 0], X[y_kmeans==2, 1], s=100, c= "green", label= "Cluster 3")

plt.scatter(X[y_kmeans==3, 0], X[y_kmeans==3, 1], s=100, c= "cyan", label= "Cluster 4")

plt.scatter(X[y_kmeans==4, 0], X[y_kmeans==4, 1], s=100, c= "magenta", label= "Cluster 5")
```

```python
plt.scatter(kmeans.cluster_centers_[:, 0],
            kmeans.cluster_centers_[:, 1],
            s= 200, c="yellow", label= "Centroids")
```

```python
plt.figure(figsize= (10,8))

plt.scatter(X[y_kmeans==0,0], X[y_kmeans==0,1], s=100, c= "red", label= "Cluster 1")

plt.scatter(X[y_kmeans==1, 0], X[y_kmeans==1, 1], s=100, c= "blue", label= "Cluster 2")

plt.scatter(X[y_kmeans==2, 0], X[y_kmeans==2, 1], s=100, c= "green", label= "Cluster 3")

plt.scatter(X[y_kmeans==3, 0], X[y_kmeans==3, 1], s=100, c= "cyan", label= "Cluster 4")

plt.scatter(X[y_kmeans==4, 0], X[y_kmeans==4, 1], s=100, c= "magenta", label= "Cluster 5")

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c='black', label='Centroids')

plt.title("Customer seg")

plt.xlabel("annual income")

plt.ylabel("spending score")

plt.legend()

plt.show()
```
