```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('D:/Besant Technologies/Python_Data_Science/Datasets/FISHMORPH_Family_Dataset.csv')
df.head()
```

```python
# The dataset contains columns like MBl (Mandible length), BEl (Eye length), etc.
# The target variable is 'Family'.
features = ['MBl', 'BEl', 'BLs', 'RMl'] # Using a subset of numeric features for simplicity
X = df[features]
y = df['Family']
```

```python
# 2. Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42) # 70% training and 30% test
```

```python
# 3. Create and train the Decision Tree Classifier
# We use 'entropy' as the criterion for information gain calculation
clf = DecisionTreeClassifier(criterion="entropy", max_depth=7)

# Train the Classifier
clf = clf.fit(X_train, y_train)
```

```python
# 4. Predict the response for the test dataset
y_pred = clf.predict(X_test)
```

```python
# 5. Model Evaluation
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```

```python
# 6. Visualize the Decision Tree
plt.figure(figsize=(20,10))
plot_tree(clf, feature_names=features, class_names=clf.classes_.tolist(), filled=True, rounded=True)
plt.show()
```
