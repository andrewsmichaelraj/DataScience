```python

import random
```

```python
print("Enter the lower bound number: ")
lower = int(input())
print("Enter the upper bound number: ")
upper = int(input())
if upper <= lower:
    print("Your upper bound should be greater than",lower,"please reenter the upper bound number")
    upper = int(input())
number = random.randrange(lower, upper)
#print(number)
total_chances = 7
print("You have only",total_chances,"chances to guess the number")
for i in range(1,9):
    if i == 8:
        print("You have exhausted all your guessess")
        break
    else:
        print("Guess the number attempt #",i)
        j = int(input())
        if (j < lower) or (j > upper):
            print("Your guess should be between",lower," & ",upper," values only")
        else:
            if j == number:
                print("You guessed is correctly")
                break
            elif j >= number:
                print("your guess is too high")
            elif j <= number:
                print("your guess is too low")
i = i + 1
```
